Convierte el precio en un `float`

```python
l=[]
for i in range(0,4622):
    l.append(float(df['item_price'][i].replace('$','')))
print(l)
```

Cuanto fue la ganancia total en el periodo del dataset?

```python
import numpy as np
q = []
for j in range (0,4622): 
    q.append(df['quantity'][j])
lxq = np.multiply(l,q)
print(sum(lxq))
```

Cual fue el promedio de ganancia por orden?

```python
prom = sum(lxq)/max(df.order_id)
prom
```
